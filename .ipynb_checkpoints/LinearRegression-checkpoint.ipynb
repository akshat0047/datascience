```python
def gradient_descent(points, some_b, some_m, iterations, learning_rate):
    b, m = some_b, some_m
    for i in range(iterations):
        b, m = step_gradient(points, b, m, learning_rate)
        
    return [b, m]
```

```python
def step_gradient(points, b, m, learning_rate):
    b_curve = 0
    m_curve = 0
    
    N = len(points)
    
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        
        # Taking partial derivaives of equation with respect to b and m
        b_curve += -(2/N) * (y - ((m * x) + b)) 
        m_curve += -(2/N) * x * (y - ((m * x) + b))
        
    new_b = b - (learning_rate * b_curve)
    new_m = m - (learning_rate * m_curve)
    
    return [new_b, new_m]
```
